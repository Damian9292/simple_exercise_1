Otwórz plik wiersz.txt i wyświetl jego treść, poprzedzając każdą linię stringiem "Wiersz <NUMER WIERSZA>: ".

Podpowiedź:

Pętli for użyj do przeczytania pliku linia po linii, a oprócz tego stwórz zmienną, która przechowa numer linii.

Numer linii możesz wstawić do stringa za pomocą metody .format(...) albo składni f-string.

```python
with open('wiersz.txt', 'r', encoding= "utf-8") as file:
    row = 0
    for line in file:
            row += 1
            print(f'Wiersz numer {row}: {line}')
```
